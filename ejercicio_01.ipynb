## Ejercicio 2

Cree un una matriz A con ```np.array```

$$\begin{bmatrix} 
	1 & 3 & 5 \\
	11 & 9 & 7\\
	13 & 15 & 17 \\
	\end{bmatrix}$$

Luego de eso utilize el metodo ```np.sort``` con  el argumento ```axis =1``` or ```axis=0```

A que conclusiones llegan? 

````python
import numpy as np # aquí importo la libreria y le asigno un homomimo para trabajar más rapido
a=np.array ([[1,3,5],[11,9,7],[13,15,17]])
print ("0. Matriz Original ")
print (a)
b=np.sort(a)# en esta línea de código igualo b a, posterior a ello la función sort ordena todo el contenido de la matriz.
print ("1. Matriz Ordenada ")
print (b)
c=np.sort(a, axis=0)
print ("2. Este es el valor de la función axis=0 de toda la matriz: ") # en esta línea de código igualo c a, posterior a ello la función axis=0 me trae los valores por columna.
print (c)
d=np.sort(a, axis=1)
print ("3. Este es el valor de la función axis=1 de toda la matriz: ") # en esta línea de código igualo b a, posterior a ello la función axis=1 me trae los valores por filas.
print (d)
````

## Ejercicio 3

Cree una matriz de 3x3 B, como la siguiente:

$$\begin{bmatrix} 
	9 & 1 & 2 \\
	1 & 9 & 5\\
	2 & 5 & 9 \\
	\end{bmatrix}$$

1. Convierta la matriz a un vector de 1x9
2. Ordene de de menor a mayor el vector
3. Aplique un filtro para extraer solo los numeros menores que 9

````python
import numpy as np # aquí importo la libreria y le asigno un homomimo para trabajar más rapido
a=np.array ([[9,1,2],[1,9,5],[2,5,9]])
print ("0. Matriz Original ")
print (a)
b=a.flatten()
print("1. Conversión de una Matriz a vector: ",b)
c=np.sort(b)
print ("2. Vector organizado con la función sort: ",c)
d=c[c<9]
print ("3. Vector con los valores menores a 9: ",d)
````


Cree una matriz de 3x3 C, como la siguiente:

$$\begin{bmatrix} 
	1 & 4 & 7 \\
	2 & 5 & 8\\
	3 & 6 & 9 \\
	\end{bmatrix}$$

1. Calcule la media por cada columna
2. Calcule la media por cada fila
3. Obtenga la media por columnas dandole el doble de peso a la columna con mayores valores de la variable
4. Calcule el percentil 25 por cada columna

````python
import numpy as np # aquí importo la libreria y le asigno un homomimo para trabajar más rapido
a=np.array ([[1,4,7],[2,5,8],[3,6,9]])
print ("0. Matriz Original ")
print (a)
print ("1. Media por columna axis=0")
b=np.nanmean(a,axis=0)
print (b)
print ("2. Media por fila axis=1")
c=np.nanmean(a,axis=1)
print (c)
print ("3. Media por la columna mayor incremnetada")
d=np.average(a,axis=0, weights=(1,1,2))# la columna 3 derecha a izquierda tiene lo mayores valores.
print (d)
e=np.percentile(a,25,axis=0)
print ("4. Percentil 25 por cada una de las columnas")
print (e)
````
